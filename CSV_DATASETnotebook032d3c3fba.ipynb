```python
import os
import numpy as np
import pandas as pd
from PIL import Image

def images_to_dataframe(data_path, image_size=(150, 150)):
    """
    Converts images from a folder into a pandas DataFrame.

    Parameters:
        data_path (str): Path to the folder containing class subfolders.
        image_size (tuple): Resize dimensions for the images (default is (150, 150)).
    
    Returns:
        pd.DataFrame: DataFrame containing flattened image data and labels.
    """
    data = []
    labels = []
    
    for label in os.listdir(data_path):
        label_path = os.path.join(data_path, label)
        if os.path.isdir(label_path):
            for image_file in os.listdir(label_path):
                image_path = os.path.join(label_path, image_file)
                try:
                    with Image.open(image_path) as img:
                        # Convert to grayscale and resize
                        img = img.convert('L').resize(image_size)
                        # Flatten the image data
                        img_array = np.array(img).flatten()
                        data.append(img_array)
                        labels.append(label)
                except Exception as e:
                    print(f"Error processing {image_path}: {e}")
    
    # Create DataFrame
    df = pd.DataFrame(data)
    df['label'] = labels
    
    return df

# Define paths for train and test datasets
train_data_path = "/kaggle/input/csv-file/BRAIN_TUMOR/training"
test_data_path = "/kaggle/input/csv-file/BRAIN_TUMOR/testing"

# Convert train and test images to DataFrames
train_df = images_to_dataframe(train_data_path)
test_df = images_to_dataframe(test_data_path)

# Add a 'dataset' column to distinguish between train and test
train_df['dataset'] = 'train'
test_df['dataset'] = 'test'

# Combine train and test DataFrames
combined_df = pd.concat([train_df, test_df], ignore_index=True)

# Save DataFrame to CSV in Kaggle working directory
output_csv_path = "/kaggle/working/mri_images_data.csv"
combined_df.to_csv(output_csv_path, index=False)

print(f"Combined DataFrame created with shape: {combined_df.shape}")
print(f"CSV file saved at: {output_csv_path}")
```
